```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('titanic_2.csv')
df
```

```python
#make some EDA in the data
df.describe()
```

```python
#check nulls values
df.isnull()
```

```python
#drop null values in Age column
df['Age'].dropna(inplace=True)
df
```

```python
#draw distplot in Age column
sns.distplot(df["Age"])
```

```python
#draw histogram in Age column.
sns.histplot(df["Age"])
```

```python
#show the number of male and female in sex column
df['Sex'].count()
```

```python
#show the number of male and female for each class in pclass column
df.groupby('Sex').count()
```

```python
#create a new column called person it's value(male,female,children(if age <16 ))
def fun_add(row):
    if row['Age'] < 16:
        return 'children'
    elif row['Sex'] == 'male':
        return 'male'
    else:
        return 'female'

df['Person'] = df.apply(fun_add, axis=1)
df.head()
```

```python
#apply step 8 again
#show the number of male and female for each class in pclass column
df.groupby('Sex').count()
```

```python
#drop null values from Cabin column
df['Cabin'].dropna(inplace=True)
df
```

```python
# make column cabin with first letter in it
df.sort_values(by=['Cabin'],inplace=True)
df
```

```python
#show the number of each cabin
df['Cabin'].unique()
```

```python
#show the number of each cabin
df['Cabin'].value_counts()
```

```python
#show the number of each element in Embarked column
df['Embarked'].value_counts()
```

```python
#create new column called alone which consist of parch column with sibsp column
df['alone'] = df['Parch'] + df['SibSp']
df
```

```python
#replace the values in survivor column from 0,1 to 'no' and 'yes'.
df['Survived'] = df['Survived'].replace({0: 'no', 1: 'yes'})
df
```

```python
# plot the bar plot between pclass and survived.
sns.barplot(x=df['Pclass'],y=df['Survived'])
```
